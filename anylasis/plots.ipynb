```python
import numpy as np

import matplotlib.pyplot as plt

def create_plot(file,title,ylim=(0,10),figsize=(8, 5)):
    data = np.loadtxt(file, delimiter=',', skiprows=1)
    nr_zapytania = data[:, 0]
    czas_zapytania = data[:, 1]


    plt.figure(figsize=figsize)
    if ylim is not None:
        plt.ylim(ylim)
    plt.plot(nr_zapytania,czas_zapytania)
    plt.xlabel('Nr zapytania')
    plt.ylabel('Czas zapytania')
    plt.title(title)
    plt.grid(True)
    plt.show()

def create_mean_plot(file,title,ylim=(0,10),figsize=(8, 5),group_size=100):
    data = np.loadtxt(file, delimiter=',', skiprows=1)
    nr_zapytania = data[:, 0]
    czas_zapytania = data[:, 1]
    num_full_groups = len(czas_zapytania)
    # Grupowanie pełnych grup
    czasy_pelne = czas_zapytania[:num_full_groups * group_size]
    srednie_czasy = np.mean(czasy_pelne.reshape(-1, group_size), axis=1)

    # Jeśli zostały dane po pełnych grupach, dodajemy średnią dla końcówki
    if len(czas_zapytania) % group_size != 0:
        ostatnia_grupa = czas_zapytania[num_full_groups * group_size:]
        srednie_czasy = np.append(srednie_czasy, np.mean(ostatnia_grupa))

    # Numeracja grup
    srednie_nr = np.arange(1, len(srednie_czasy) + 1)


    plt.figure(figsize=(8, 5))
    if ylim is not None:
        plt.ylim(ylim)
    plt.plot(srednie_nr, srednie_czasy)
    plt.xlabel('Nr zapytania')
    plt.ylabel('Czas zapytania')
    plot_title = title+", uśrednione co "+str(group_size)+" zapytań"
    plt.title(plot_title)
    plt.grid(True)
    plt.show()
```

## HTTP



### n_times

```python
title = 'HTTP'

create_plot('../output/http_small_session.csv', title,ylim=(0,4))
create_mean_plot('../output/http_small_session.csv', title ,ylim=(0,4), group_size=100)
```

### Single

```python
file = '../output/http_single.csv'
ylim = (2, 8)
title = 'HTTP'

create_plot(file, title,ylim)
create_mean_plot(file, title,ylim, group_size=100)
```

### File

```python
file = '../output/http_big.csv'
ylim = None
title = 'HTTP'

create_plot(file, title,ylim)
create_mean_plot(file, title ,ylim, group_size=10)
```
